```python
plt.plot([5,10,15,20,25], "rH:") # format syntax "[color][marker][solidline]"

# color code => r->red c->cyan k->black b->blue m->megenta w->white g->green
# marker code => .->point ,->pixel 0->circle ^->triangle up marker <->left >->right triangle 1->tridown
# s->square *->for start x->for x D->diamand H-> hexagone h-> hexagone
# line code => --> solid line, -- -> dashline, -. -> dash dot : -> dotted line
```

```python
import numpy as np
a = np.arange(0, 3, 0.5)
plt.plot(a**2, "g^-",a**3, "r*--", a**4, "yD-.", a**5, "bs:")
```
