## data visualization

```python
# read the csv by pandas
import pandas as pd
df_population_raw = pd.read_csv(r"C:\Users\JiyaUlla\Desktop\DataSetsForDtaScience\population_total.csv")
```

```python
df_population_raw
```

```python
df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot(index='year', columns='country', values='population')
```

```python
df_pivot = df_pivot[['United States', 'India', 'China', 'Brazil', 'Indonesia']]
```

```python
df_pivot
```

```python
df_pivot.plot(kind='line')
```

```python
df_pivot.plot(kind='line', xlabel="Year", ylabel="Population", title="Population from 1955- 2000")
```

```python
df_pivot.plot(kind='line', xlabel="Year", ylabel="Population", title="Population from 1955- 2000", figsize=(16, 8))
```
